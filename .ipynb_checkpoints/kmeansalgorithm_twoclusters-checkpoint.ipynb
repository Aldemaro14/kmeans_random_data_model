```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
#Generate Data - two means
mean1 = [np.random.randint(50), np.random.randint(50)]
mean2 = [np.random.randint(50), np.random.randint(50)]

#Make diagonal covarience
cov = [[100, 0], [0, 100]]

x1, y1 = np.random.multivariate_normal(mean1, cov, 100).T
x2, y2 = np.random.multivariate_normal(mean2, cov, 100).T

x = np.append(x1, x2)
y = np.append(y1, y2)

plt.plot(x, y, 'x')
plt.axis('equal')
plt.show
```

```python
X = np.array(list(zip(x, y))) #for Python 2.7 zip returns a tupple, for python 3.x you need to add list() 
```

```python
#Make KMeans model
kmeans = KMeans(n_clusters = 2)
```

```python
#Fit the model to the data
kmeans.fit(X)
```

```python
centroids = kmeans.cluster_centers_
labels = kmeans.labels_
print(labels)
print('\n')
print(centroids)
```

```python
colors = ["g.", "r."]

for i in range(len(X)):
    plt.plot(X[i][0], X[i][1], colors[labels[i]], markersize = 10)
    
plt.scatter(centroids[:, 0], centroids[:, 1], marker = "X", s = 150, zorder = 10)
plt.show()
```
